```python
# score.txt에 있는 점수를  읽어온 후 합계와 평균 구한 후 출력
f = open("./score.txt", "r", encoding="utf-8")
scores = []
for num in f:
    scores.append(int(num))
print(f"평균: {sum(scores) / len(scores)}")
```

```python
# sample.txt 읽은 후 합계와 평균 구하기
# result.txt 작성(합계, 평균 기록)
f = open("./sample.txt", "r", encoding="utf-8")
scores = []
for num in f:
    scores.append(int(num))
f.close()
f = open("./result.txt", "w", encoding="utf-8")
f.write(f"합계: {sum(scores)}\n평균: {sum(scores) / len(scores)}")
f.close()
```

```python
# info.txt 읽기

# BMI 지수 계산 후 화면에 출력
# [출력결과]
# 아라, 57, 47
# 이름: 아라
# 몸무게: 57
# 키: 158
# BMI: 33.156(bmi = weight / (height / 100) ** 2)
# 결과: 과체중

f = open("./info.txt", "r", encoding="utf-8")
for i in f:
     name, weight, height = i.split(", ")
     bmi = int(weight) / (int(height) / 100) ** 2

     if bmi >= 25:
          result = "과체중"
     elif bmi >= 18.5:
          result = "정상체중"
     else:
          result = "저체중"
          
     print(i, end="")
     print(f"이름: {name}")
     print(f"몸무게: {weight}")
     print(f"키: {int(height)}")
     print(f"BMI: {bmi}")
     print(f"결과: {result}\n")
     
    
f.close()

```

```python
while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료"))
    if no == 1:
        # origin.txt 읽어 암호화 후 encryption.txt로 저장
        f1 = open("./origin.txt", "r", encoding="utf-8")
        content = f1.read()
        f2 = open("./encryption.txt", "w", encoding="utf-8")
        for i in content:
            f2.write(chr(ord(i) + 100))
    elif no == 2:
        # encryption.txt를 읽은 후 원래 문자대로 화면 출력
        f1 = open("./encryption.txt", "r", encoding="utf-8")
        content = f1.read()
        for i in content:
            print(chr(ord(i) - 100), end="")
    else:
        break
```

```python
# close() 자동으로 호출해줌
with open("./origin.txt", "r", encoding="utf-8") as f1:
    print(f1.read())
```
