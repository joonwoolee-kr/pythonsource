#### 화면 조작
- find_element(By.ID)
- find_element(By.CLASS_NAME)
- find_element(By.XPATH)
- find_element(By.SELECTOR)

- click(): 클릭
- send_keys(): 키 입력

```python
from chrome import set_chrome_driver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = set_chrome_driver()
driver.get("https://www.naver.com")

# 검색창에 검색어 입력 후 엔터
# 검색 요소 찾기
element = driver.find_element(By.ID, "query")
# print(element)
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)

# 검색 후 페이지 소스 가져오기
print(driver.page_source)
# 파싱: selenium or BeautifulSoup

time.sleep(3)

driver.quit()
```

```python
from chrome import set_chrome_driver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = set_chrome_driver()
driver.get("https://www.danawa.com/")
time.sleep(2)

# 로그인 클릭
login_btn = driver.find_element(By.CLASS_NAME, "btn_user--login")
login_btn.send_keys(Keys.ENTER)

time.sleep(1)

# 검색창에 검색어 입력 후 엔터
# 검색 요소 찾기
login_id = driver.find_element(By.ID, "danawa-member-login-input-id")
login_pwd = driver.find_element(By.ID, "danawa-member-login-input-pwd")
login_id.send_keys("")
login_pwd.send_keys("")
login_pwd.send_keys(Keys.ENTER)

time.sleep(3)

driver.quit()
```
